## Dataset Summary

For this lab, I selected a **video game sales dataset from Kaggle** that contains sales data for video games across multiple platforms and regions worldwide.

### Features Included in the Dataset
- **Rank** - Overall sales ranking of the game
- **Name** - Title of the video game
- **Platform** - Gaming platform the title was released on (Wii, PS2, Xbox 360, etc.)
- **Year** - Release year
- **Genre** - Type of game (Sports, Action, Role-Playing, Racing, etc.)
- **Publisher** - Company that published the game
- **NA_Sales** - Sales in North America (millions)
- **EU_Sales** - Sales in Europe (millions)
- **JP_Sales** - Sales in Japan (millions)
- **Other_Sales** - Sales in other regions (millions)
- **Global_Sales** - Total worldwide sales (millions)

## Research Questions

For this analysis, I will explore the following questions using Pandas and data visualizations:

1. Which video game genres have generated the highest total global sales?
2. Which gaming platforms have generated the highest total global sales?
3. Which publishers have generated the highest total global sales?
4. How do regional gaming preferences differ between North America, Europe, and Japan?
5. How have video game sales changed over time by release year?

```python
import pandas as pd
import matplotlib.pyplot as plt

games_df = pd.read_csv('../Lecture/Week 1/assets/vgsales.csv')

games_df.head()
```

## Question 1

**Which video game genres have generated the highest total global sales?**

```python
genre_sales = games_df.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)

genre_sales
```

```python
genre_sales.plot(kind='bar', figsize=(10,6))

plt.title('Total Global Sales by Video Game Genre')
plt.xlabel('Genre')
plt.ylabel('Global Sales (Millions)')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

**Answer:**

Based on the analysis, **Action games** generated the highest total global sales at approximately **1,751 million units**, making them the most commercially successful genre in the dataset. Sports and Shooter games followed behind, while Strategy games had the lowest total global sales. This suggests that fast-paced, broadly appealing genres tended to perform best in the global gaming market.

## Question 2

**Which gaming platforms have generated the highest total global sales?**

```python
platform_sales = games_df.groupby('Platform')['Global_Sales'].sum().sort_values(ascending=False)

top_platforms = platform_sales.head(10)

top_platforms
```

```python
top_platforms = platform_sales.head(10)

top_platforms.plot(kind='bar', figsize=(12,6))

plt.title('Top 10 Gaming Platforms by Total Global Sales')
plt.xlabel('Platform')
plt.ylabel('Global Sales (Millions)')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

**Answer:**  
The PlayStation 2 (PS2) generated the highest total global video game sales at approximately 1,255.64 million units, making it the most successful gaming platform in the dataset. The Xbox 360 and PlayStation 3 followed closely behind, with 979.96 million and 957.84 million global sales respectively. Nintendo platforms also performed strongly, with the Wii and Nintendo DS ranking among the top five. These results show that Sony, Microsoft, and Nintendo dominated the global gaming market during the time periods represented in the dataset.

# Question 3

**Which publishers have generated the highest total global sales?**

```python
publisher_sales = games_df.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False)

top_publishers = publisher_sales.head(10)

top_publishers
```

```python
publisher_sales = games_df.groupby('Publisher')['Global_Sales'].sum().sort_values(ascending=False)

top_publishers = publisher_sales.head(10)

top_publishers.plot(kind='bar', figsize=(12,6))

plt.title('Top 10 Video Game Publishers by Total Global Sales')
plt.xlabel('Publisher')
plt.ylabel('Global Sales (Millions)')
plt.xticks(rotation=60, ha='right')
plt.tight_layout()

plt.show()
```

**Answer:**  
Nintendo generated the highest total global video game sales by a significant margin, with approximately 1,787.66 million units sold worldwide. Electronic Arts ranked second with about 1,110.32 million global sales, followed by Activision with 727.46 million. Sony Computer Entertainment and Ubisoft also performed strongly among the top publishers. These results suggest that Nintendo has had exceptional long-term global success, likely driven by iconic franchises such as Mario, Pokémon, and Wii Sports, while major third-party publishers like Electronic Arts and Activision also maintained strong market influence.

# Question 4

**How do regional gaming preferences differ between North America, Europe, and Japan?**

```python
regional_genre_sales = games_df.groupby('Genre')[['NA_Sales', 'EU_Sales', 'JP_Sales']].sum()

regional_genre_sales
```

```python
regional_genre_sales = games_df.groupby('Genre')[['NA_Sales', 'EU_Sales', 'JP_Sales']].sum()

regional_genre_sales.plot(kind='bar', figsize=(14,7))

plt.title('Regional Video Game Genre Preferences')
plt.xlabel('Genre')
plt.ylabel('Sales (Millions)')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

# Question 5

**How have video game sales changed over time by release year?**

```python
yearly_sales = games_df.groupby('Year')['Global_Sales'].sum().sort_index()

yearly_sales
```

```python
yearly_sales.plot(kind='line', figsize=(12,6), marker='o')

plt.title('Global Video Game Sales Over Time')
plt.xlabel('Release Year')
plt.ylabel('Global Sales (Millions)')
plt.grid(True)

plt.show()
```

**Answer:**  
Global video game sales increased steadily over time, with modest sales in the 1980s and early 1990s before experiencing rapid growth in the late 1990s and 2000s. Sales peaked around 2008 at approximately 678.90 million units, indicating the height of the traditional console gaming era. After this peak, global sales began to decline, likely due to shifts in the gaming industry such as mobile gaming growth, digital distribution changes, and evolving consumer preferences. This trend shows the rise, peak, and decline of physical video game sales over time.
