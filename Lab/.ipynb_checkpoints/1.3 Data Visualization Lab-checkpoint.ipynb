# Intro to Data Visualization Lab
## Part 1: Feature Visualizations and Typical Values

```python
import matplotlib.pyplot as plt

df['bpm'].hist()
plt.title('Distribution of BPM')
plt.xlabel('BPM')
plt.ylabel('Number of Songs')
plt.show()

df['bpm'].describe()
```

### BPM Analysis
The BPM distribution appears approximately centered around 120 BPM, with most songs falling between 100 BPM and 140 BPM. Based on this distribution, a likely BPM range for a guaranteed smash hit would be **100 - 140 BPM**, with a typical value near **120 BPM**.

```python
df['key'].value_counts().plot(kind='bar')
plt.title('Distribution of Musical Keys')
plt.xlabel('Key')
plt.ylabel('Number of Songs')
plt.show()

df['key'].value_counts()
```

### Key Analysis
The most common musical key in the Spotify dataset is **C#**, with 120 songs using this key. Since this appears more frequently than any other key, **C# would be the most likely choice for a guaranteed smash hit**.

```python
df['mode'].value_counts().plot(kind='bar')
plt.title('Distribution of Musical Modes')
plt.xlabel('Mode')
plt.ylabel('Number of Songs')
plt.show()

df['mode'].value_counts()
```

### Mode Analysis
The most common musical mode in the dataset is **Major**, appearing in 550 songs compared to 403 songs in Minor mode. Based on this distribution, a guaranteed smash hit would most likely use **Major mode**.

```python
df['danceability_%'].hist()
plt.title('Distribution of Danceability')
plt.xlabel('Danceability %')
plt.ylabel('Number of Songs')
plt.show()

df['danceability_%'].describe()
```

### Danceability Analysis
The danceability distribution is centered around 69%, with most songs falling between 57% and 78%. Based on this distribution, a likely danceability range for a guaranteed smash hit would be **57% - 78%**, with an ideal value near **69%**.

```python
df['energy_%'].hist()
plt.title('Distribution of Energy')
plt.xlabel('Energy %')
plt.ylabel('Number of Songs')
plt.show()

df['energy_%'].describe()
```

### Energy Analysis
The energy distribution is centered around 66%, with most songs falling between 53% and 77%. This suggests that a guaranteed smash hit would likely have an energy level between **53% and 77%**, with an ideal target around **66%**.

```python
df['speechiness_%'].hist()
plt.title('Distribution of Speechiness')
plt.xlabel('Speechiness %')
plt.ylabel('Number of Songs')
plt.show()

df['speechiness_%'].describe()
```

### Speechiness Analysis
The speechiness distribution is heavily concentrated at lower values, with most songs containing very little spoken-word content. The median speechiness is 6%, and 50% of songs fall between 4% and 11%. Based on this distribution, a guaranteed smash hit would likely have a speechiness value between **4% and 11%**, with an ideal target around **6%**.

```python
df['acousticness_%'].hist()
plt.title('Distribution of Acousticness')
plt.xlabel('Acousticness %')
plt.ylabel('Number of Songs')
plt.show()

df['acousticness_%'].describe()
```

### Acousticness Analysis
The acousticness distribution shows that most popular songs lean toward lower acoustic values, meaning they are generally more electronically produced rather than heavily acoustic. The median acousticness is 18%, with 50% of songs falling between 6% and 43%. Based on this distribution, a guaranteed smash hit would likely have an acousticness value between **6% and 43%**, with an ideal target around **18%**.

```python
df['artist(s)_name'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Most Frequent Artists')
plt.xlabel('Artist')
plt.ylabel('Number of Songs')
plt.xticks(rotation=45)
plt.show()

df['artist(s)_name'].value_counts().head(10)
```

### Artist Analysis
The artist distribution shows that some artists appear much more frequently in the dataset than others. Taylor Swift appears the most often with 34 songs, followed by The Weeknd with 22 songs. Based on this frequency, Taylor Swift represents the most common artist profile associated with popular songs in this dataset.

```python
engineered_hit = df[
    (df['bpm'] >= 100) & (df['bpm'] <= 140) &
    (df['key'] == 'C#') &
    (df['mode'] == 'Major') &
    (df['danceability_%'] >= 57) & (df['danceability_%'] <= 78) &
    (df['energy_%'] >= 53) & (df['energy_%'] <= 77) &
    (df['speechiness_%'] >= 4) & (df['speechiness_%'] <= 11) &
    (df['acousticness_%'] >= 6) & (df['acousticness_%'] <= 43)
]

engineered_hit[['track_name', 'artist(s)_name']]
```

## Engineered Smash Hit Results

Using the criteria identified from the dataset:

- BPM: 100 - 140
- Key: C#
- Mode: Major
- Danceability: 57% - 78%
- Energy: 53% - 77%
- Speechiness: 4% - 11%
- Acousticness: 6% - 43%

Two songs in the Spotify dataset matched all of the engineered hit criteria:

1. **Need To Know** by **Doja Cat**
2. **Ginseng Strip 2002** by **Yung Lean**

This suggests that the engineered formula successfully identified real songs that match the characteristics of popular music in the dataset.
