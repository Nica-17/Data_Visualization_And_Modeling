```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Lecture/Week 1/assets/titanic_passengers.csv')
df.head()
```

```python
pd.crosstab(df['Pclass'], df['Survived'])
```

```python
pd.crosstab(df['Pclass'], df['Survived'], normalize=True) * 100
```

```python
cross_tab = pd.crosstab(df['Pclass'], df['Survived'])

cross_tab.plot(kind='bar', stacked=True)

plt.title('Titanic Survival by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Number of Passengers')
plt.legend(['Perished', 'Survived'])

plt.show()
```

```python
pd.crosstab(df['Pclass'], df['Survived'], normalize='index') * 100
```
