```python
import pandas as pd

titanic_df = pd.read_csv('assets/titanic_passengers.csv')

titanic_df.head()
```

```python
third_class = titanic_df[titanic_df['Pclass'] == 3]

third_class.head()
```

```python
third_class.groupby('Survived')['Age'].mean()
```

```python
titanic_df['Fare in CAD'] = titanic_df['Fare'] * 1.70

titanic_df[['Fare', 'Fare in CAD']].head()
```

```python
pd.crosstab(titanic_df['Pclass'], titanic_df['Survived'])
```

Based on the crosstab results, passengers in 3rd class were the least likely to survive the Titanic voyage, with 372 deaths compared to 119 survivors.

```python
titanic_df['Port_name'] = titanic_df['Embarked'].map({
    'S': 'Southampton',
    'C': 'Cherbourg',
    'Q': 'Queenstown'
})

titanic_df[['Embarked', 'Port_name']].head()
```
