```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../Week 2/assets/titanic_passengers.csv')

df.head()
```

```python
df.groupby('Pclass')['Age'].agg(['mean', 'median'])
```

```python
plt.figure(figsize=(10,6))

sns.histplot(df[df['Pclass'] == 1]['Age'], bins=20, kde=True, label='1st Class', alpha=0.5)
sns.histplot(df[df['Pclass'] == 2]['Age'], bins=20, kde=True, label='2nd Class', alpha=0.5)
sns.histplot(df[df['Pclass'] == 3]['Age'], bins=20, kde=True, label='3rd Class', alpha=0.5)

plt.title('Passenger Age Distribution by Ticket Class')
plt.xlabel('Age')
plt.ylabel('Number of Passengers')
plt.legend()

plt.show()
```

```python
df['Pclass_cat'] = df['Pclass'].astype(str)
```

```python
plt.figure(figsize=(8,6))

sns.boxplot(data=df, x='Pclass_cat', y='Age', order=['1', '2', '3'])

plt.title('Passenger Age Distribution by Ticket Class')
plt.xlabel('Passenger Class')
plt.ylabel('Age')

plt.show()
```

The overlaid histogram makes it easier to observe the overall shape and overlap of the age distributions between passenger classes, showing that third-class passengers were concentrated at younger ages while first-class passengers skewed older.

The side-by-side boxplots make it easier to compare medians, interquartile ranges (IQR), and outliers between the passenger classes.

First-class passengers had the highest center of age distribution, with both the highest mean and median ages, indicating an older passenger population. Their age distribution also had the widest spread, suggesting greater variability in ages.

Second-class passengers had a moderate center and spread, falling between first and third class in average age.

Third-class passengers had the youngest center, with the lowest mean and median ages, and their distribution was more concentrated among younger adults and children, though some older outliers were present.
